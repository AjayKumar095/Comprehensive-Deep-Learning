<h3>Q1: Basic Components of a Digital Image and Representation in a Computer</h3>

<p><b>Answer: -</b></p>
A digital image is essentially a matrix of pixel values, each representing a tiny portion of the image. The main components of a digital image are:

- Pixels: The smallest individual unit in an image, representing color or intensity.
- Resolution: The number of pixels across the width and height of an image, often denoted as width × height (e.g., 1920 × 1080).
- Color Depth: Specifies how many bits are used to represent each color or intensity level, influencing the range of colors or shades available.

In a computer, a digital image is represented as an array of pixel values. Each pixel can hold color or grayscale information, depending on the type of image.

Differences between Grayscale and Color Images:
- Grayscale Images: Each pixel represents a single intensity value, ranging from 0 (black) to 255 (white) in an 8-bit representation. Grayscale images have no color information and are simpler to process.
- Color Images: Each pixel holds multiple intensity values, typically for red, green, and blue (RGB) channels. Each channel has a range from 0 to 255, creating a variety of colors when combined. Color images are three-dimensional matrices (width × height × channels).

<h3>Q2: Convolutional Neural Networks (CNNs) and Their Role in Image Processing.</h3>
<p><b>Answer: -</b></p>

Convolutional Neural Networks (CNNs) are specialized types of neural networks designed primarily for processing structured grid data, like images. CNNs excel in extracting patterns and features (such as edges, textures, and shapes) from images, making them ideal for tasks like image classification, object detection, and segmentation.

Key Advantages of CNNs over Traditional Neural Networks:
1. Spatial Hierarchy of Features: CNNs use multiple layers to automatically capture spatial hierarchies, detecting low-level features (like edges) in early layers and higher-level features (like shapes or objects) in deeper layers.
2. Parameter Efficiency: By using convolutional layers with shared weights (filters), CNNs reduce the number of parameters, making training more computationally efficient.
3. Translation Invariance: The use of convolutional filters helps CNNs detect features regardless of their position in the image, enhancing their ability to generalize.

<h3>Q3: Convolutional Layers, Filters, Padding, and Strides in CNNs.</h3>
<p><b>Answer: -</b></p>

- Convolutional Layers: Convolutional layers are the core of a CNN. They apply a set of filters (or kernels) across an image to detect specific patterns or features.

- Filters: Filters are small matrices (e.g., 3×3 or 5×5) that slide over the input image, performing element-wise multiplication and summation to produce a single output value. Filters are designed to detect features such as edges, textures, or patterns in the input image. Each filter captures a specific feature, with multiple filters used to capture various aspects of the input data.

Padding and Strides:
- Padding: Padding adds extra pixels around the border of an image to maintain the spatial dimensions of the output. Common padding options are:
- Valid Padding: No extra pixels are added, leading to a smaller output.
- Same Padding: Zero-padding is added so that the output size remains the same as the input.
- Strides: Strides determine how many pixels the filter moves after each application. A stride of 1 means the filter moves one pixel at a time, whereas a stride of 2 moves the filter two pixels. Larger strides reduce the output size and computation.

Impact on Output Size: Using padding can preserve the original dimensions of the input, whereas strides control how much the output size is reduced with each convolution. Together, they impact the overall size of the feature map generated by the convolutional layer.

<h3>Q4: Purpose of Pooling Layers in CNNs and Comparison of Max Pooling vs. Average Pooling.</h3>
<p><b>Answer: -</b></p>

Pooling layers in CNNs reduce the spatial dimensions of the feature maps, helping to:

- Reduce computational load by decreasing the number of parameters.
- Control overfitting by condensing features and making the model less sensitive to positional variations.
- Retain the most important features, allowing CNNs to focus on critical parts of the data.

Max Pooling vs. Average Pooling:

- Max Pooling: Takes the maximum value from a region (e.g., 2×2) in the feature map. This operation retains the most prominent features, making it ideal for capturing distinct features.

- Average Pooling: Calculates the average value of each region in the feature map. This operation gives a more generalized representation of the data, which can be useful in some contexts but is often less effective for capturing distinct features than max pooling.
